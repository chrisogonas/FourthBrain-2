2.  How does the Prophet Algorithm differ from an LSTM?

`Prophet is specifically designed for business time series prediction. Prophet requires less hyperparameter tuning as it is specifically designed to detect patterns in business time series.`

`LSTMs are more powerful in their approach to learning from sequential data and time series. They also perform better when learning from massive datasets with complex patterns. Unlike Prophet, they make no specific assumptions about the data such as time series stationarity. However, LSTMs are difficult to interpret in their behavior. Further, they require careful hyperparameter tuning is required to achieve good results.`

3.  Why does an LSTM have poor performance against ARIMA and Profit for Time Series?

`The AutoRegressive Integrated Moving Average (ARIMA) model uses the last p time series values as features, where p is a hyperparameter that needs to be determined when designing the model. The model can also use past errors. These aspects make it possible to add as features the most recent q errors the model made, where q is another hyperparameter. Importantly, ARIMA model standardizes time series such that the model becomes independent from seasonal or temporary trends. This is the stationarity of time series.`

`Long short-term memory (LSTM) cells are used in recurrent neural networks to predict the future from sequences of variable lengths. Unlike ARIMA, LSTMs are not restricted to time series. LSTM cells use gate functions to learn the important parts of the sequence seen so far and forget the less important ones.`

`In conclusion, the stationarity aspect of ARIMAs give them an advantage in performance over time series data when compared to LSTMs that take input sequences without any averaging.`

5. Stationarity, Seasonality, and Cyclicality

What is stationarity?

`It is the statistical properties of a process generating a time series not changing over time. However, this does not mean that the series does not change over time; instead, it means that the way it changes does not itself change over time.`

What is seasonality?

`It is the time series component that describes the seasonal fluctuations i.e. the component is the function that models seasonal trends.`

Why Is Stationarity Important in Time Series Forecasting?
